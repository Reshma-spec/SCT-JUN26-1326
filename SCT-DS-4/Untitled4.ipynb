```python
import pandas as pd

df = pd.read_csv("accidents_data/US_Accidents_March23.csv")

print(df.head())
print(df.shape)
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

top_states = df['State'].value_counts().head(10)

plt.figure(figsize=(10,6))
top_states.plot(kind='bar')

plt.title('Top 10 States by Number of Accidents')
plt.xlabel('State')
plt.ylabel('Accident Count')

plt.savefig('top_states.png')
plt.show()
```

```python
weather = df['Weather_Condition'].value_counts().head(10)

plt.figure(figsize=(10,6))
weather.plot(kind='bar')

plt.title('Top Weather Conditions During Accidents')
plt.xlabel('Weather Condition')
plt.ylabel('Count')

plt.savefig('weather_conditions.png')
plt.show()
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

df['Hour'] = df['Start_Time'].dt.hour

hourly = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(12,6))
hourly.plot(kind='line')

plt.title('Accidents by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Number of Accidents')

plt.savefig('accidents_by_hour.png')
plt.show()
```

```python
sample = df.sample(10000, random_state=42)

plt.figure(figsize=(10,6))

plt.scatter(
    sample['Start_Lng'],
    sample['Start_Lat'],
    s=1
)

plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.savefig('accident_hotspots.png')
plt.show()
```

```python
road_conditions = {
    'Bump': df['Bump'].sum(),
    'Crossing': df['Crossing'].sum(),
    'Junction': df['Junction'].sum(),
    'Traffic_Signal': df['Traffic_Signal'].sum(),
    'Stop': df['Stop'].sum()
}

plt.figure(figsize=(8,5))

plt.bar(
    road_conditions.keys(),
    road_conditions.values()
)

plt.title('Road Features Near Accidents')
plt.xticks(rotation=45)

plt.savefig('road_conditions.png')
plt.show()
```
